```python
import pandas as pd
import numpy as np
from pathlib import Path
import nltk
import random
from collections import defaultdict
```

```python
from nltk.tokenize import WordPunctTokenizer
from nltk.corpus import stopwords
import string
from collections import defaultdict
from nltk.stem.snowball import EnglishStemmer
```

```python
ROOT = Path().absolute()
txt_mus = open(ROOT/"data/musicas_short.txt", "r", encoding='utf-8')
lyrics = txt_mus.read()
```

```python
songs_list = lyrics.split("\n\nXXXXXXXXXX\n\n")
songs_list[9200].split("\n---\n")[0].split("\n>> ") + [songs_list[9200].split("\n---\n")[1]]
```

```python
all_lyrics = []
i = 0
for song in songs_list:
    lyrics1 = song.split("\n---\n")
    lyrics = lyrics1[0].split("\n>> ") + [lyrics1[1]]
    all_lyrics.append(lyrics)
```

```python
nltk.download('stopwords')
```

```python
size = int(len(all_lyrics)*0.7)  # training set size
train_sample_ids = random.sample(range(0, len(all_lyrics)), size)
train_sample_ids.sort()
train_sample_ids
```

```python
test_sample_ids = []
for i in range(0, len(all_lyrics)):
    if (i not in train_sample_ids):
        test_sample_ids.append(i)

len(test_sample_ids) + len(train_sample_ids)
```

```python
train_set = []
for idx in train_sample_ids:
    train_set.append(all_lyrics[idx])

# train_set
```

```python
swu = stopwords.words('english') + list(string.punctuation)
swu
```

```python
stemmer = EnglishStemmer()
stemmer.stem('making')
```

```python
letra_tokenizada = []
tokenizador = WordPunctTokenizer()
a = train_set[1567][2].lower()
for token in tokenizador.tokenize(a):
    if (token not in swu):
        letra_tokenizada.append(token)
# letra_tokenizada
```

```python
clean_lyrics = {'genero':[], 'musica':[]}
tokenizador = WordPunctTokenizer()
for song in train_set:
    current_song = []
    song_lower = song[2].lower()
    genre = song[1]
    for token in tokenizador.tokenize(song_lower):
        if (token not in swu):
            current_song.append(token)
    clean_lyrics['genero'].append(genre)
    clean_lyrics['musica'].append(current_song)
# clean_lyrics['musica']
```

```python
lyrics_df = pd.DataFrame(data = clean_lyrics)
lyrics_df.groupby('genero').count().sort_values(by = ['musica'], \
    ascending=False)[:5]
```

```python
dicio_saco_de_palavras = {}
for genero in lyrics_df['genero'].unique():
    cur_words = defaultdict(lambda: 0)
    for musica in lyrics_df[lyrics_df['genero'] == genero]['musica']:
        for word in musica:
            cur_words[word] += 1
    dicio_saco_de_palavras[genero] = cur_words
dicio_saco_de_palavras
```

```python
pop_df = pd.DataFrame.from_dict(data = dicio_saco_de_palavras['pop'], orient = 'index', columns=['frequency'])
pop_df.sort_values(by = 'frequency', ascending=False, inplace=True)

soul_df = pd.DataFrame.from_dict(data = dicio_saco_de_palavras['soul'], orient = 'index', columns=['frequency'])
soul_df.sort_values(by = 'frequency', ascending=False, inplace=True)

indie_df = pd.DataFrame.from_dict(data = dicio_saco_de_palavras['indie'], orient = 'index', columns=['frequency'])
indie_df.sort_values(by = 'frequency', ascending=False, inplace=True)

eletro_df = pd.DataFrame.from_dict(data = dicio_saco_de_palavras['eletronica'], orient = 'index', columns=['frequency'])
eletro_df.sort_values(by = 'frequency', ascending=False, inplace=True)

rock_df = pd.DataFrame.from_dict(data = dicio_saco_de_palavras['rock and roll'], orient = 'index', columns=['frequency'])
rock_df.sort_values(by = 'frequency', ascending=False, inplace=True)


gospel_df = pd.DataFrame.from_dict(data = dicio_saco_de_palavras['gospel'], orient = 'index', columns=['frequency'])
gospel_df.sort_values(by = 'frequency', ascending=False, inplace=True)
```

```python
from gensim.test.utils import common_texts
from gensim.models import Word2Vec

model = Word2Vec(sentences=common_texts, vector_size=100, window=5, min_count=1, workers=4)
```

```python
model.train([["hello", "world"]], total_examples=1, epochs=1)
```

```python
v1 = model.wv['computer']
v2 = model.wv['system']
v3 = model.wv['graph']
v4 = model.wv['trees']
dist_v1v2 = np.linalg.norm(v1-v2)
dist_v1v3 = np.linalg.norm(v1-v3)
dist_v1v4 = np.linalg.norm(v1-v4)
print(f'Distância entre "computer" e "system": {dist_v1v2}\nDistância entre "computer" e "graph": {dist_v1v3}\nDistância entre "computer" e "trees": {dist_v1v4}')
```
